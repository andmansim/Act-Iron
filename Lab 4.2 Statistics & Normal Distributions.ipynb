#### Find the mean and the standard deviation of the temperature of each state.

*Hint: Use functions from Data Manipulation lesson*

```python
# your answer here
def estado(temp, est):
    lista = []
    for i in range(len(temp)):
        if temp.State[i] == est:
            lista.append(temp.Temperature[i])
    return lista

ny = estado(temp, 'New York')
c = estado(temp, 'Connecticut')
nj = estado(temp, 'New Jersey')
media_ny = np.mean(ny)
desv_ny = np.std(ny)
media_c = np.mean(c)
desv_c = np.std(c)
media_nj = np.mean(nj)
desv_nj = np.std(nj)
print('New York: media, ' + str(media_ny) + ' desviación típica ' + str(desv_ny))
print('Connecticut: media, ' + str(media_c) + ' desviación típica ' + str(desv_c))
print('New Jersey: media, ' + str(media_nj) + ' desviación típica ' + str(desv_nj))

```

#### Find the employees who's title is associate and the salary above 55?

```python
# your answer here
def empleados(employee, titulo):
    lista = []
    for i in range(len(employee)):
        if employee.Title[i] == titulo:
            if  employee.Salary[i] > 55:
                lista.append(employee.Name[i])
                
    return lista
a = empleados(employee, 'analyst')
b = empleados(employee, 'associate')
c = empleados(employee, 'VP')
print('Analyst ' + str(a), 'Associate ' + str(b), 'VP ' + str(c))
```

```python
# your answer here

def ns (employee, nombre):
    sal = []
    anio = []
    for i in range(len(employee)):
            if employee.Department[i] == nombre :
                sal.append(employee.Salary[i])
                anio.append(employee.Years[i])
    return sal, anio
sal1, an1 = ns(employee, 'IT')
sal2, an2 = ns(employee, 'HR')
sal3, an3 = ns(employee, 'Sales')
print('Maximo IT: salario '+ str(max(sal1)) + ' años '+ str(max(an1)))
print('Maximo HR: salario '+ str(max(sal2)) + ' años ' +str(max(an2)))
print('Maximo Sales: salario '+ str(max(sal3)) + ' años '+ str(max(an3)))
print('Mínimo IT: salario '+ str(min(sal1)) + ' años ' +str(min(an1)))
print('Minimo HR: salario '+ str(min(sal2)) + ' años '+ str(min(an2)))
print('Minimo Sales: salario '+ str(min(sal3)) + ' años ' +str(min(an3)))
print('Media IT: salario '+ str(np.mean(sal1)) + ' años ' +str(np.mean(an1)))
print('Media HR: salario '+ str(np.mean(sal2)) + ' años ' +str(np.mean(an2)))
print('Media Sales: salario '+ str(np.mean(sal3)) + ' años ' +str(np.mean(an3)))
```
